```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sensor1 = pd.read_csv('../data/sensor_data/sensor_data_part-1.csv')
sensor1.tail()
```

```python
# take 1 booking id as example
example_trip = sensor1[sensor1.bookingID == 0]
example_trip = example_trip.sort_values(by='second')
example_trip.head()

# matplotlib multiple lineplot
plt.figure(figsize=(15, 5))
plt.plot(example_trip['second'], example_trip['acceleration_x'], label='acceleration_x')
plt.plot(example_trip['second'], example_trip['acceleration_y'], label='acceleration_y')
plt.plot(example_trip['second'], example_trip['acceleration_z'], label='acceleration_z')
plt.plot(example_trip['second'], example_trip['Speed'], label='speed')
plt.legend()
plt.show()
```

```python
# matplotlib multiple lineplot
plt.figure(figsize=(15, 5))
plt.plot(example_trip['second'], example_trip['gyro_x'], label='acceleration_x')
plt.plot(example_trip['second'], example_trip['gyro_y'], label='acceleration_y')
plt.plot(example_trip['second'], example_trip['gyro_z'], label='acceleration_z')
plt.legend()
plt.show()
```
